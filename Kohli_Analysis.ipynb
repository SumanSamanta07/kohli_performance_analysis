```python
def find_centuries(x):
    if x>100:
        return "OG"
    else:
        return "NOOB"

df["Centuries"]=df["Runs"].apply(find_centuries)
```

```python
sum=df["Runs"].sum()
balls=df["BF"].sum()
avg=df["SR"].sum()/df["SR"].count()
print(sum)
print(balls)
print(avg)
```

```python
# Number of matches he has played at different positions
positions = df["Pos"].unique()
print(positions)

df["Pos"]=df["Pos"].map({
    3.0:"Batting at 3",
    4.0:"Batting at 4",
    2.0:"Batting at 2",
    1.0:"Batting at 1",
    7.0:"Batting at 7",
    5.0:"Batting at 5",
    6.0:"Batting at 6"
})
```

```python
pos_counts=df["Pos"].value_counts()
print(pos_counts)
print(type(pos_counts))
```

```python
import matplotlib.pyplot as plt
pos_counts_values = pos_counts.values
pos_counts_labels = pos_counts.index

fig = plt.figure(figsize=(10,7))
plt.pie(pos_counts_values,labels=pos_counts_labels)
plt.show()
```

```python
# Total sixes scored in different positions

pos_6 = df.groupby("Pos")["6s"].sum()
print(pos_6)
```

```python
def show_pie_plot(df,key):
    counts=df[key].value_counts()
    print(counts)
    count_values=counts.values
    count_labels=counts.index
    fig = plt.figure(figsize=(10,7))
    plt.pie(count_values,labels=count_labels)
    plt.show()

# Number of matches he has played with different teams

show_pie_plot(df,"Opposition")  
```
